```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots we will create
```

```python

#Read cvs file

df = pd.read_csv(r'C:\Users\easyc\Documents\DATA\DATAANALYTICS\DATAPROJECTS\Correlation in Python\movies.csv')
```

```python
#Look at the dataa

df.head()
```

```python
# Dealing with missin values

df.dropna(inplace=True)
```

```python
#looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))## if we want to round the output we use  round(pct_missing*100)
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
#changing DF type
df['budget'] = df['budget'].astype('int64') # when the columns contain Nan values, we have to use Int64 with the I caps.

df['votes'] = df['votes'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')

df['year'] = df['year'].astype('str')
```

```python
df
```

```python
# Creating the correct year column

date1 = df['released'].str.split(',', expand=True)

date1[1].astype('str')
```

```python
date1[1].str.split(' ', expand=True)

year2 = date1[1].str.split(' ', expand=True)

df['yearcorrect'] = year2[1]
```

```python
df['yearcorrect'].describe()
```

```python
##Ordering the Data by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# High correlation between budget and gross
# Heatmap visualization


correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

# add title
plt.title('Correlation Matrix for Numeric features')

# labels

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at Company
# Company is not numeric, but we can create a numeric representation of it.
# for example: Fox is numer 1, Marvel studios is 2, Lucasfilm is 3, etc, etc, etc


df.head()
```

```python
# create a numeric representation for all the columns
df2 = df
df_numerized = df2

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
                                  
df_numerized   
```

```python
df
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

# add title
plt.title('Correlation Matrix for Numeric features')

# labels

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# We want to see what have the better correlation quickly
# we are using unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)

sorted_pairs
```

```python
sorted_descending = corr_pairs.sort_values(ascending=True)
high_corr = sorted_descending[(sorted_pairs) > 0.4]
high_corr
```

```python
# Votes and Budget have the highest correlation to gross

# Company has low correlation
```
